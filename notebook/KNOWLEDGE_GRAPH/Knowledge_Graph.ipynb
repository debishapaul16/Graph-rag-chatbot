#INTRODUCTION

#What is Knowledge Graph:A Knowledge Graph is a structured, graph-based representation of entities and the relationships between them

#Nodes : These are the entities or objects in the graph, such as a person, company, product, or concept. Each node can have properties (attributes) that provide more details about the entity.

#Edges : These are the connections or relationships between nodes. For example, an edge might represent that "Alice works at Company X".Here "works at" stands as the relationship connecting the two nodes.

#Properties/Labels : Nodes and edges can have properties or labels that describe their characteristics or the type of relationship. For example, a node for a person might have properties like name and birthdate, while an edge might have a label like "employed by" or "located in".

#Triples:A Triple is the basic building block of a Knowledge Graph.(Entity 1, Relationship, Entity 2),Subject  → Predicate → Object.

Knowledge graphs can be saved in many formats. Primary formats include

NetworkX : In-memory representation, easy to work with for small and medium sized graphs.It can be used for running algorithms, testing ideas or visualizing smaller graphs during prototyping

```python
#CREATION OF GRAPH OBJECT
import networkx as nx

data=[("RAG","uses","Knowledge Graph"),
      ("Knowledge Graph","contains","Triples"),
      ("NetworkX","is_a","Python Library"),
      ("Data Science","uses","Python")
     ]

G = nx.DiGraph()#creates a empty directed graph

for source, relation, target in data:
    G.add_edge(source, target, relation=relation)
```

```python
#VISUALIZATION OF GRAPH
import matplotlib.pyplot as plt
plt.figure(figsize=(8,6))


pos = nx.spring_layout(#breaks each and every dataset for example ("RAG","uses","Knowledge Graph") and then breaks it according to the position
    G,
    k=1.5,      # controls spacing between each node
    seed=42     # fixed layout so that it doesnot change everytime we run
)               

nx.draw(
    G,#graph object
    pos,#position for each value
    with_labels=True,#label the nodes
    node_size=2500,#node size
    font_size=10,
    arrows=True
)

edge_labels = nx.get_edge_attributes(G, 'relation')#return the relation from the pos
nx.draw_networkx_edge_labels(
    G,
    pos,
    edge_labels=edge_labels
)


plt.show()#visualizing the graph created
```

```python
#FOR SHOWING THE RELAIONSHIPS

for u, v, data in G.edges(data=True): #so here basically the graph data is segregated and from there each entity is chosen and broken down into three parts as per the triples
                                               #node1-->relation-->node2
    print(f"{u} --{data['relation']}--> {v}")
```

```python
#for querying the knowledge_graph:Knowledge Graph Management
for u, v, data in G.edges(data=True):
    if data["relation"] == "is_a" and v == "Python Library":
        print(u)
```

```python
#KNOWLEDGE GRAPH MANAGEMENT

#add relationship to the graph
G.add_edge(
    "Charlie",
    "Amazon",
    relation="works_at"
)
#update relationship
G["Charlie"]["Amazon"]["relation"] = "interns_at"

#find neighbour

list(G.neighbors("Charlie"))

#delete relationship
G.remove_edge(
    "Data Science",
    "Python"
)

```

```python
plt.figure(figsize=(8,6))


pos = nx.spring_layout(#breaks each and every dataset for example ("RAG","uses","Knowledge Graph") and then breaks it according to the position
    G,
    k=1.5,      # controls spacing between each node
    seed=42     # fixed layout so that it doesnot change everytime we run
)               

nx.draw(
    G,#graph object
    pos,#position for each value
    with_labels=True,#label the nodes
    node_size=2500,#node size
    font_size=10,
    arrows=True
)

edge_labels = nx.get_edge_attributes(G, 'relation')#return the relation from the pos
nx.draw_networkx_edge_labels(
    G,
    pos,
    edge_labels=edge_labels
)


plt.show()#visualizing the graph created
```
